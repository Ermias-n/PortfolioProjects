```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Set the default style for Matplotlib to 'ggplot'
plt.style.use('ggplot')

# Set the default figure size for Matplotlib
plt.rcParams['figure.figsize'] = (12, 8)

%matplotlib inline

# Read a CSV file into a Pandas DataFrame
df = pd.read_csv(r'C:\Users\ERMI\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Iterate through each column in the DataFrame
for col in df.columns:
    # Calculate the percentage of missing values in the current column
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Get the data types of each column in the DataFrame 'df'
df.dtypes
```

```python
# Remove rows with non-finite values in 'budget' and 'gross' columns
df.dropna(subset=['budget', 'gross'], inplace=True)
```

```python
# Converts 'budget' and 'gross' columns to 'int64', 
# but handle non-finite values (NaN or inf) beforehand.

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Extracts the year from the 'released' column and stores it in the 'yearcorrect' column as a string.
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
df
```

```python
# Sort DataFrame 'df' by the 'gross' column in descending order.

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Create a regression plot using Seaborn to visualize the relationship between 'budget' and 'gross' columns in DataFrame 'df'

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"purple"}, line_kws={"color":"green"})
```

```python
# Calculate the Pearson correlation matrix for all numeric columns in the DataFrame 'df'
df.corr(numeric_only=True, method='pearson')
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
